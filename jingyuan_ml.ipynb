```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = sns.load_dataset("penguins")

# Display the first few rows of the dataframe
print(df.head(6))
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Handling missing values - Example: drop rows with any missing values
df_cleaned = df.dropna()
```

```python
# Check for duplicates
print(df_cleaned.duplicated().sum())
```

```python
# Removing duplicates if any
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# Check data types
print(df_cleaned.dtypes)
```

```python
# Convert a categorical feature from object to category
df_cleaned['species'] = df_cleaned['species'].astype('category')
df_cleaned['island'] = df_cleaned['island'].astype('category')
df_cleaned['sex'] = df_cleaned['sex'].astype('category')
```

```python
# Summary statistics
print(df_cleaned.describe())
```

```python
# Examine the distribution of features
# Histogram of flipper length
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['flipper_length_mm'], kde=True)
plt.title('Distribution of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot 
# bill length vs. bill depth
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Bill Length vs. Bill Depth by Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend(title='Species')
plt.show()
```

```python
# Box plot 
# body mass by species
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned, x='species', y='body_mass_g')
plt.title('Body Mass by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```
